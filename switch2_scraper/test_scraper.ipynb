```python
from scraper.client import NintendoLifeClient
from scraper.games import GameListScraper
from scraper.reviews import ReviewScraper

client = NintendoLifeClient(delay=0.5)
game_scraper = GameListScraper(client)
review_scraper = ReviewScraper(client, max_per_game=5)

# 1. Get games
games = game_scraper.scrape_browse_page(1)

# 2. Test first game with reviews
test_game = games[2]  # Pokémon Pokopia (known to have reviews)
print(f"Testing {test_game.name}...")

review_links = review_scraper.scrape_review_links(test_game)
print(f"Found {len(review_links)} review links")

# 3. Get full review data
full_reviews = []
for review in review_links:
    full_review = review_scraper.scrape_full_review(review)
    full_reviews.append(full_review)

# 4. Show results
for i, review in enumerate(full_reviews, 1):
    print(f"\nReview {i}:")
    print(f"Game: {review.game_name}")
    print(f"Score: {review.score}")
    print(f"Text: {review.review_text[:150]}...")
    print(f"URL: {review.review_url}")
```
